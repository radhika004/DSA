```python
class Node:
  
  def __init__(self,data):
    self.data = data
    self.next = None
    
class LinkedList:
  
  def __init__(self):
    self.head = None
    
  
  def insert_values(self):
    arr = []
    n = int(input("Enter size of the linkedlist"))
    for i in range(n):
      x = int(input("Enter Element"))
      arr.append(x)
    
    return arr
  
  
  def create(self,arr):
    
    if len(arr)==0:
      return
      
    self.head = Node(arr[0])
    last = self.head
    for i in range(1,len(arr)):
      t = Node(arr[i])
      last.next = t
      last = t
    last.next =self.head
    
  def display(self,p):
    if self.head is None:
        return

    p = self.head
    while True:
        print(p.data, end=" ")
        p = p.next
        if p == self.head:
          break
      
    
    
l = LinkedList()
r = l.insert_values()

l.create(r)
l.display(l.head)

    
      
      


      
      
    
    
  
    
    
```
