```python
class Node:
  
  def __init__(self,data):
    self.data = data
    self.next = None
    
class LinkedList:
  
  def __init__(self):
    self.head = None
    
  
  def insert_values(self):
    arr = []
    n = int(input("Enter size of the linkedlist"))
    for i in range(n):
      x = int(input("Enter Element"))
      arr.append(x)
    
    return arr
  
  
  def create(self,arr):
    
    if len(arr)==0:
      return
      
    self.head = Node(arr[0])
    last = self.head
    for i in range(1,len(arr)):
      t = Node(arr[i])
      last.next = t
      last = t
    last.next =self.head
    
  def display(self,p):
    if self.head is None:
        return

    p = self.head
    while True:
        print(p.data, end=" ")
        p = p.next
        if p == self.head:
          break
    print()
        
        
  def insert(self,p,pos,k):      
    t = Node(k)
    
    if pos == 0:
      if (p == None):
        p = t
        p.next = p
      
      else:
        while p.next != self.head:
          p = p.next
        p.next = t
        t.next = p
        p = t
        
    else:
      for i in range(pos-1):
        p = p.next
      t = Node(k)
      t.next = p.next
      p.next = t
  
  def delete_node(self, p, pos):

    if self.head is None:
        return

    if pos == 1:

        if self.head.next == self.head:
            self.head = None
            return

        while p.next != self.head:
            p = p.next

        p.next = self.head.next
        self.head = self.head.next

    else:
        for i in range(pos - 2):
            p = p.next

        q = p.next
        p.next = q.next
        q.next = None
      
    
        
        
      
    
    
  
    
    
l = LinkedList()
r = l.insert_values()

l.create(r)
l.display(l.head)

l.insert(l.head,2,8)
l.display(l.head)

l.delete_node(l.head,2)
l.display(l.head)


    
      
      


      
      
    
    
  
    
    
```

```python
# implementing doubly linked list 

class Node:
  def __init__(self,data):
        self.prev = None
        self.data = data
        self.next = None

class Doubly_LL:
  def __init__(self):
    self.first = None
    
  def create(self,A,n):
    self.first.data = Node(A[0])
    self.first.prev=self.first.next = None
    last = self.first
    
    for i in range(n):
      t = Node(A[i])
      t.next = last.next
      t.prev =  last
      last.next = t
      last = t;
    
    def display (self,p):
      while p :
        print(p.data,end = " ")
        p = p.next
      print()
      
    
    def length(self,p):
      leng = 0
      while(p):
        leng = leng + 1
        p = p.next
      
      return leng
    
  
  
l = Doubly_LL()
A = [10, 20, 30, 40, 50]

l.create(A, 5)
l.display(l.first)
print(l.length(l.first))
  
        
    
    
    
    
```
