```python
class Node:
  
  def __init__(self,data):
    self.data = data
    self.next = None
  
class LinkedList:
  def __init__(self):
    self.first = None
    self.second = None
    self.third = None
  
  def input_values(self):

    arr =  []
    n = int(input("Enter the number of elements of the array"))
    for i in range(n):
      x = int(input("Enter element:"))
      
      arr.append(x)
    
    return arr    
  
  def create(self,arr):
    if len(arr)==0:
      self.first = None
      
    self.first = Node(arr[0])
    last = self.first
    for i in range(1,len(arr)):
      t = Node(arr[i])
      last.next = t
      last = t
    
  def display(self,p):
    # while p is not None:
    #   print(p.data,end=" ")
    #   p=p.next
    
    if p is not None:
      print(p.data,end=" ")
      self.display(p.next)
    print()
      
  def count_elements(self,p):
    cout = 0
    
    while p is not None :
      cout = cout+1
      p = p.next
    return cout
      
  def search_element(self,p,d):
    while p is not None:
      if p.data == d:
        print("Yes")
        return
      p=p.next
    print("No")
    
  def maximum_element(self,p):
    maxi = 0
    
    while p is not None:
      if p.data>maxi:
        maxi = p.data
      p = p.next
    return maxi
  
  def smallest(self,p):
    mini = p.data
    while p is not None:
      if p.data < mini:
        mini = p.data
      p  = p.next
    return mini
      
  
      
  def addition(self,p):
    add = 0
    while p is not None:
      add = add + p.data
      p =  p.next
    return add
  
  
  def move_to_head(self,p,key):
    q = None
    while p is not None:
      if key==p.data:
        if q is not None:
          q.next=p.next
          p.next = self.first
          self.first = p
        return
      q=p
      p=p.next
      
  
  # insert new element to the linked list
  
  def insert_before_first(self,key):
    q = Node(key)
    q.next = self.first
    self.first = q
    
  def insert_at_pos(self,p,pos,x):
    q = Node(x)
    for i in range(pos-1):
      p = p.next
    
    q.next = p.next
    p.next = q
        
  
  def insert_last(self,val):
    q = Node(val)
    
    if self.first is  None:
      self.first = last = q
      return
    
    last = self.first
    
    while last.next is not None:
      last = last.next
    
    last.next = q
    
    
  def delete_first_node(self):
    if self.first is None:
      return 
    
    q = self.first
    self.first = self.first.next
    del q
    
  def delete_at_pos(self,pos):
    p = self.first
    if self.first is None:
      return
    
    if pos ==1:
      q = self.first
      self.first = self.first.next
      del q
      return
    
    p = self.first
    for i in range(pos-2):
      if p is None:
        return
      p = p.next
      
    if p is  None or p.next is None:
      return
    
    q = p.next
    p.next = q.next
    del q
    
    
    
  def is_sorted(self,p):
    x = float('-inf') 
    while p is not None:
      if x>p.data:
        return False
      x = p.data
      p =p.next
    return True
    
  def remove_duplicates(self):
    if self.first is None:
      return
    
    p = self.first
    q = self.first.next
    
    while  q is not None:
      if p.data != q.data:
        p = q
        q = q.next
      else:
        p.next = q.next
        q = p.next
   
  def rev_list(self,p):
    a = []
    while p is not None:
      a.append(p.data)
      p=p.next
    
    p = self.first
    for i in range(len(a)-1,-1,-1):
        p.data = a[i]
        p = p.next
    
  
  def rev_3ptr(self):
    p = self.first
    q = None
    r = None
    
    while p is not None:
      r = q
      q = p
      p = p.next
      q.next = r
    
    
    self.first = q
    
  def recursive_reverse(self,p,q):
    
    if p is not None:
      self.recursive_reverse(q,p.next)
      p.next = q
    else:
      self.first = q
      
    
  #  new linked list creation
  def create1(self,arr):
    if len(arr)==0:
      self.second = None
      
    self.second = Node(arr[0])
    last = self.second
    for i in range(1,len(arr)):
      t = Node(arr[i])
      last.next = t
      last = t

  def concat_list(self,p):
    if p is None:
      self.first = self.second
      return
    while p.next != None:
      p = p.next
    
    p.next = self.second
      
      
      
  def mergeing(self):
    
    if self.first is None:
        self.third = self.second
        return

    if self.second is None:
        self.third = self.first
        return
    
    
    if self.first.data < self.second.data:
      self.third = last = self.first
      self.first = self.first.next
      last.next = None
    
    else:
      self.third = last = self.second
      self.second = self.second.next
      last.next = None
      
    while self.first!=None and self.second!=None:
      if self.first.data < self.second.data:
        last.next = self.first
        last = self.first
        self.first = self.first.next
        last.next = None
        
      else:
        last.next = self.second
        last = self.second
        self.second = self.second.next
        last.next = None
        
    if self.first is not None:
      last.next = self.first
    else:
      last.next = self.second
    
    
    
  
    
l = LinkedList()
r =l.input_values()
l.create(r)
l.display(l.first)
l.search_element(l.first,10)
d = l.maximum_element(l.first)
print("maximum element:",d)

s = l.smallest(l.first)
print("smallest elemet =",s)
c = l.count_elements(l.first)
a = l.addition(l.first)
print("addition of all the elements:",a)

print("Total no. of elemenst:",c,end="\n")
      
# l.move_to_head(l.first,2)
# l.display(l.first)

# l.insert_before_first(23)
# l.display(l.first)

# l.insert_at_pos(l.first,2,45)
# l.display(l.first)

# l.insert_last(20)
# l.display(l.first)

# l.delete_first_node()
# l.display(l.first)

# l.delete_at_pos(3)
# l.display(l.first)

# l.is_sorted(l.first)

# l.remove_duplicates()
# l.display(l.first)


l.rev_list(l.first)
l.display(l.first)

l.recursive_reverse(None,l.first)
l.display(l.first)


```
