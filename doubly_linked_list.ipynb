```python
class Node:
    def __init__(self, data):
        self.prev = None
        self.data = data
        self.next = None


class Doubly_LL:
    def __init__(self):
        self.first = None

    def create(self, A, n):
        self.first = Node(A[0])
        last = self.first

        for i in range(1, n):
            t = Node(A[i])
            t.next = None
            t.prev = last
            last.next = t
            last = t

    def display(self, p):
        while p:
            print(p.data, end=" ")
            p = p.next
        print()

    def length(self, p):
        leng = 0
        while p:
            leng += 1
            p = p.next
        return leng
    
    def insert(self,p,n,pos):
        
        if pos<0 or pos>self.length(p):
            return
        
        if pos == 0:
           t = Node(n)
           t.prev = None
           t.next = p
           p.prev = t
           p = t
        else:
            t = Node(n)
            for i in range(pos-1):
                p = p.next
            t.next = p.next
            t.prev = p
            if p.next:
                p.next.prev = t
            p.next = t


l = Doubly_LL()
A = [10, 20, 30, 40, 50]

l.create(A, 5)
l.display(l.first)
print(l.length(l.first))

l.insert(l.first,34,2)
l.display(l.first)

```

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution(object):
    def swapPairs(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        dummy = ListNode(0)
        dummy.next = head

        prev = dummy

        while prev.next and prev.next.next:
            first = prev.next
            second = first.next

            first.next = second.next
            second.next = first
            prev.next = second

            prev = first

        return dummy.next
```
