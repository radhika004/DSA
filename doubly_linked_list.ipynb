```python
class Node:
    def __init__(self, data):
        self.prev = None
        self.data = data
        self.next = None


class Doubly_LL:
    def __init__(self):
        self.first = None

    def create(self, A, n):
        self.first = Node(A[0])
        last = self.first

        for i in range(1, n):
            t = Node(A[i])
            t.next = None
            t.prev = last
            last.next = t
            last = t

    def display(self, p):
        while p:
            print(p.data, end=" ")
            p = p.next
        print()

    def length(self, p):
        leng = 0
        while p:
            leng += 1
            p = p.next
        return leng


l = Doubly_LL()
A = [10, 20, 30, 40, 50]

l.create(A, 5)
l.display(l.first)
print(l.length(l.first))
```
