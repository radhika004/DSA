```python
# use dictionaries 
n = int(input("Enter size of array: "))
a = []
for i in range(0, n):
    a.append(int(input()))

# Create an empty dictionary map
hash_map = {}

for i in range(0, len(a)):
    # .get(key, default) returns 0 if the number isn't in the map yet
    hash_map[a[i]] = hash_map.get(a[i], 0) + 1

number = int(input("Enter number to find its duplicate: "))
print(f"{number} occurred {hash_map.get(number, 0)} times")
```

```python
# Linked list implementation 

class Node:
  def __init__(self,data):
    self.data = data
    self.next = None
    
  first = None
  
def create(A):
    global first
    first = Node(A[0])
    last = first
    for i in range(1,len(A)):
      t = Node(A[i])
      last.next = t
      last = t
def display(p):
    while p is not None:
      print(p.data,end=" ")
      p = p.next
    print()

def sum_list(s):
  total_sum = 0
  while s is not None:
    total_sum = total_sum + s.data
    s = s.next
  return total_sum

def count_list(c):
  cout = 0
  while c is not None:
    cout = cout + 1
    c = c.next
  return cout
    
def searching(d,target):
  nod = 0
  while d is not None:
    if d.data == target:
      print("yes,Its is at the node L:",nod)
    
    d = d.next
    nod = nod + 1

def maximum(p):
  maxi = p.data
  while p is not None:
    if p.data >maxi:
      maxi = p.data
    p = p.next
  return maxi

def linear_search(p,key):
  global first
  q = None
  while p is not None:
    if key==p.data:
      if p == first:
        return p
      q.next = p.next
      p.next = first
      first = p
      return p
    q=p
    p=p.next
  return None
    
if __name__=="__main__":
  A = [3,5,7,8,10]
  create(A)
  display(first)
  result = sum_list(first)
  print(result)
  no_elements = count_list(first)
  print(no_elements)
  searching(first,8)
  maximun_number = maximum(first)
  print(maximun_number)
  linear_search(first,10)
  display(first)
```

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        c = []
        for i in range(len(s)):
            for j in range(i+1,len(s)+1):
                b = s[i:j]
                c.append(b)
        ans = 0
        for sub in c:
            if len(sub) == len(set(sub)):
                ans=max(ans,len(sub))
        
        return ans

        
            

```
